```python
#Imorting
import re
import nltk

#Added this because the tagger is not being recongnized
import nltk
#nltk.download('averaged_perceptron_tagger_eng')
#nltk.download('punkt')
import os
nltk.data.path.append(os.path.expanduser('~/nltk_data'))



# Improved normalization function
# Removes all punctuation except periods (for sentences)
# Lowercases all words and tokenizes
def tknize (a_string):
    # Handle all the string operations at one time
    clean = re.sub('[^a-zA-Z \.]', ' ', a_string).lower()
    # Create the list of sub-strings (tokens) and return it
    return nltk.tokenize.word_tokenize(clean)

# DATA
with open("C:/Users/alex/ENGL370-2025/Garcia/The_Art_of_War.txt", mode="r", encoding="utf-8") as f:
            aow = f.read()

print(aow[15000:16000])
```

```python
# Tokenizing the text
aow_ = tknize(aow)

# Printing to display the random terms
print(aow_[100:250])
```

```python
# mdg_ is our cleaned list of tokens which keeps only periods.
aow_tagged = nltk.pos_tag(aow_)

aow_tagged[15000:16000]
```

```python

# Empty out the list of adverbs. Then this is where Adverbs are going to be put in with text from 15000, to 16000 count of the text.
# 
adverbs = []
for i in aow_tagged:
    if i[1] == "RB":
        adverbs.append(i[0])
        #print(i[0]))
print(i[15000:16000])
# Give me the number count of adverbs from 15000 to 16000.
print(len(adverbs))
```

```python
# How to Read the Grammar below:
# Start with an optional (?) determiner ('DT')
# Can have any number (*) of adjectives (JJ)
# End with a noun (<NN>)
grammar = "NP: {<DT>?<JJ>*<NN>}"

# Instantiate the chunk parser
parser = nltk.RegexpParser(grammar)

# Run it on our tagged text
tree = parser.parse(mdg_tagged)

# See some results
for i in tree[0:20]:
    print(i)
```
