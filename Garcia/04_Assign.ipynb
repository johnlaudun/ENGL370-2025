```python
from collections import Counter

# Function to get most common words from nltk
def most_common_words(tokens, n=10):
    word_counts = Counter(tokens)
    return word_counts.most_common(n)

# Most common words for each text
common_words_text2 = most_common_words(tokens2)
common_words_text3 = most_common_words(tokens3)
common_words_text4 = most_common_words(tokens4)

# Print the most common words for each text
print("Most common words in Text 2:")
print(common_words_text2)

print("\nMost common words in Text 3:")
print(common_words_text3)

print("\nMost common words in Text 4:")
print(common_words_text4)
```

```python
import re

# Tokenize each text (Text 2, Text 3, Text 4)
with open("The_Inventions_by_Nikola_Tesla", "r", encoding="utf-8") as file:
    text2 = file.read()
tokens2 = re.findall(r"\b\w+(?:'\w+)?\b", text2)  # Tokenizing using regular expression

with open("Opportunities_in_Engineering.txt", "r", encoding="utf-8") as file:
    text3 = file.read()
tokens3 = re.findall(r"\b\w+(?:'\w+)?\b", text3)

with open("The_Art_of_War.txt", "r", encoding="utf-8") as file:  # Replace with actual third text file
    text4 = file.read()
tokens4 = re.findall(r"\b\w+(?:'\w+)?\b", text4)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to generate and display word cloud
def generate_word_cloud(tokens, title="Word Cloud"):
    # Convert the tokens list into a single string
    text = " ".join(tokens)
    
    # Create a word cloud object
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Plot the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")  # Remove axes for better presentation
    plt.title(title, fontsize=20)
    plt.show()

# Generate word clouds for each text (with stopwords removed if you prefer)
generate_word_cloud(tokens2, "Word Cloud for Text 2")
generate_word_cloud(tokens3, "Word Cloud for Text 3")
generate_word_cloud(tokens4, "Word Cloud for Text 4")
```
